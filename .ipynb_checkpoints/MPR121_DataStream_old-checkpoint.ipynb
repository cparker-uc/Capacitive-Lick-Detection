# MPR121 Capacitive Touch Detector Data Stream

```python
# Set Environment Variables so Adafruit Blinka knows we're using the MPR121 with FT232H
import os
os.environ["BLINKA_MPR121"] = "1"
os.environ["BLINKA_FT232H"] = "1"
import board
import busio
import adafruit_mpr121
i2c = busio.I2C(board.SCL, board.SDA)
mpr121 = adafruit_mpr121.MPR121(i2c)
```

```python
%matplotlib widget
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from collections import deque
```

```python
# Capacitance data
HISTORY_SIZE = 10
cap_data = deque([deque(maxlen=HISTORY_SIZE) for _ in range(12)])
x_time = deque([deque(maxlen=HISTORY_SIZE) for _ in range(12)])
while True:
    for chan in range(12):
        cap_data[chan].append(mpr121[chan].raw_value)
        x_time[chan].append(time.time())
    #print(x_time)
```

```python
# How many sensor samples we want to store
HISTORY_SIZE = 10

# Graph update interval (in seconds)
INTERVAL = 0.01

# Global x-axis array
x_time = deque(maxlen=HISTORY_SIZE)

# Capacitance data
cap_data = deque([deque(maxlen=HISTORY_SIZE) for _ in range(12)])

# Create new plot
fig, ax = plt.subplots(nrows=6, ncols=2, figsize=(12,7), sharex=True, sharey=True)
plt.tight_layout()

anim = animation.FuncAnimation(fig, animate, cache_frame_data=False)
```
